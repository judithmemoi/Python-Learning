```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 =pd.read_excel("Car data saved.xlsx")
data1.head()
```

```python
variables = ['Min Price (Lakh)', 'Max Price (Lakh)', 'Range (kmpl)', 'CC',
       'Seats', 'Variants', 'Type', 'Ex-Showroom Price', 'RTO', 'Insurance',
       'Other', 'Onroad Price']
```

```python
plt.figure(figsize=(15,12))
for i, var in enumerate(variables, start=1):
    plt.subplot(4, 3, i)
    sns.scatterplot(x=data1[var],y=data1['Max Price (Lakh)'], alpha=0.7)
    plt.title(f' Scatter plot of {var} vs Onroad Price')
    plt.xlabel(var)
    plt.ylabel("Target")
plt.tight_layout()
plt.show()
```

```python
# Hitograms for all independent variables
plt.figure(figsize=(15,10))
for i, var in enumerate(variables, start=1):
    plt.subplot(3, 4, i)
    sns.histplot(x=data1[var], bins=20, kde = True, color ="green")
    plt.title(f' Histogram of {var}')
    plt.xlabel(var)
    plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Create a boxplot of Range by car Type
sns.boxplot(data=data1, x="Type", y="Range (kmpl)",
            hue = "Type", palette=["blue", "Red"])
plt.title("Boxplot of Range by Car Type")
plt.show()
```

```python
selected_columns=['Min Price (Lakh)', 'Max Price (Lakh)', 'Range (kmpl)', 'CC',
       'Seats', 'Variants', 'Ex-Showroom Price', 'RTO', 'Insurance',
       'Other']
data_selected =data1[selected_columns]
data_selected.head()
```

```python
corr= data_selected.corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap="Blues")
plt.title("Heatmap of Correlation Matrix")
plt.show()
```

```python
sns.violinplot(data=data1, x="Type", y="Max Price (Lakh)",
               hue = "Type", palette=["green", "Orange"])
plt.title("Violin plot of Maxmum Price by Car Type")
plt.show()
```
