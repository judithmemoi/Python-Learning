# Data Manipulation

```python
df = df.rename(columns={
    'County/Sub County': 'SubCounty',
    'Total': 'Total_Population',
    'Farming': 'Farming_Activities',
    'Crop Production': 'Crop_Production',
    'Livestock Production': 'Livestock_Production',
    
}) #renaming columns
```

```python
df.head()
```

```python
grouped_df = df.groupby("SubCounty").sum()
print(grouped_df.head()) 
```

```python
grouped_df
```

```python
'''
In Python, aggregation is a design principle that allows one object to contain or reference another object.
It's a form of object composition where an object is composed of one or more other objects, which it uses as its parts.
Aggregation represents a "has-a" relationship, meaning that an object "has a" reference toÂ anotherÂ object.
'''
```

```python
df['Total_Agriculture'] = (df['Farming_Activities'] +
                           df['Crop_Production'] +
                           df['Livestock_Production'] +
                           df['Aquaculture'] +
                           df['Fishing'] +
                           df['Irrigation'])
print(df.head()) 
```

```python
filtered_df = df[df['Sorghum'] > 1000]
```

```python
filtered_df
```

```python
filter = (
    (df['Maize'] > 1500) & 
    (df['Camels'] > 2000) & 
    (df['Beehives'] > 2500) & 
    (df['Fish Ponds'] > 1200)
)

filtered_df =df[filter]
```
