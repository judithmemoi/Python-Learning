```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("data.xlsx")
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"]) # Convert to datetime format

#plotting the data
plt.plot(df["Date"],df["Sales"], color= "Purple")
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar chart-use to describe descrete Data
plt.bar(df["Product"],df["Sales"],color="Blue")
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
#Aggregate data by car model
Sales_data = df.groupby("Category")["Sales"].sum()
Sales_data
```

```python
# create the pie chart
plt.figure(figsize=(8,8))
plt.pie(Sales_data, labels=Sales_data.index,autopct="%1.1f%%")

plt.title("Proportion of Sales by Category")
plt.show()
```

```python
plt.hist(df["Sales"], bins=8, color="pink", edgecolor="blue")
plt.title("Histogram")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()


```

```python
plt.fill_between(df["Date"], df["Sales"], color="lightgreen", alpha=0.3) 
plt.plot(df["Date"],df["Sales"], color= "green")
plt.title("Cumulative Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```
